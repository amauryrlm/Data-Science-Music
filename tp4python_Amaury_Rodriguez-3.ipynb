## Partie 3 : Mise en place d'un modèle de Machine Learning

Pour notre modèle de ML nous avons choisit des attributs catégoritiels: 
-si une musique est explicite ou pas, en effet d'après la dernière partie nous avons vu qu'une musique est en moyenne plus populaire si elle est explicite
-son année de parution car avec le temps de plus en plus de gens ont accès a la musique la rendant populaire, cependant il y a un biais car la popularité comme comptabilisé ici est très certainement relative aux ventes de CD et il y a eu une transition du mode d'écoute (cf partie 2).

Pour les attributs numériques, j'ai pris ceux dont la valeur sont les plus élevées dans la matrice de corrélation: loudness energy et danceability ce qui parait des critères pertinents pour qu'une musique soit populaire.

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

```python
labels=df['popularity']
train=df[['loudness','explicit','energy','year','danceability']]
train.head()
labels.head()

```

```python
X_train, X_test, y_train, y_test = train_test_split(train, labels, test_size=0.66, random_state=2)

```

```python
gbr=GradientBoostingRegressor()
```

```python
gbr.fit(X_train, y_train)
```

```python
gbr.score(X_test,y_test)
```

```python
y_predgbr=gbr.predict(X_test)
mean_squared_error(y_test,y_predgbr)

cross_val_score(gbr, train , labels, cv=10)
```

la MSE est de 194 ce qui est assez élévé car la popularité va de 0 à 100 avec une moyenne de 25. En moyenne la prediction est éloigné de la réalité de environs 13 points ce qui n'est pas très précis

```python
reg= LinearRegression()
```

```python
reg.fit(X_test,y_test)
```

```python
reg.score(X_test,y_test)
```

```python
y_predreg=reg.predict(X_test)
mean_squared_error(y_test,y_predreg)
```
